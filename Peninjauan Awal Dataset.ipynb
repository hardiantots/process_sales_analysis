### Pengecekan Nilai Kategorikal Setiap Tabel

```python
pd.set_option('display.max_rows', None)

def check_categorical_value(name_df):
    list_categorical = name_df.select_dtypes(include=['object']).columns.tolist()

    for col in list_categorical:
        print(f'Kolom {col} memiliki {name_df[col].nunique()} kategori unik.')
        print(name_df[col].value_counts())
        print('-----------------------------------\n\n')
```

#### Table Sell In

```python
check_categorical_value(table_sell_in)
```

#### Table Sell Out

```python
check_categorical_value(table_sell_out)
```

#### Table CA

```python
check_categorical_value(table_CA)
```

#### Table Survey

```python
check_categorical_value(table_survey)
```
